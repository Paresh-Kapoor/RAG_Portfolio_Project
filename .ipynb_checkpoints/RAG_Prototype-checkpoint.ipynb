```python
import os
import getpass
from langchain_google_genai import ChatGoogleGenerativeAI

# 1. Securely ask for the API key via a text box
if "GOOGLE_API_KEY" not in os.environ:
    os.environ["GOOGLE_API_KEY"] = getpass.getpass("Enter your Google API key: ")

# 2. Initialize the Large Language Model (Gemini)
# We updated the model to 'gemini-1.5-flash-latest' to ensure we hit the active Free Tier quota
llm = ChatGoogleGenerativeAI(
    model="gemini-1.5-flash-latest", 
    temperature=0
)

# 3. Ask the LLM a simple test question
response = llm.invoke("What is the capital of India?")

# 4. Print the exact answer text from the AI
print(response.content)
```
